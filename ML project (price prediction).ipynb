```python
# feature engineering
fullDF1=pd.get_dummies(fullDF,drop_first=True)
```

```python
fullDF1.shape
```

```python
trainDF,testDF=train_test_split(fullDF1,train_size=0.8,random_state=2410)
```

```python
trainDF.shape
testDF.shape
```

```python
trainx=trainDF.drop('Sale_Price',axis=1)
trainy=trainDF['Sale_Price']
testx=testDF.drop('Sale_Price',axis=1)
testy=testDF['Sale_Price']
```

```python
# standardized the dataset
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler().fit(trainx)
trainx_scaled=scaler.transform(trainx)
testx_scaled=scaler.transform(testx)
```

```python
trainx_scaled
```

```python
trainx_rev=pd.DataFrame(trainx_scaled,columns=trainx.columns)
testx_rev=pd.DataFrame(testx_scaled,columns=testx.columns)
```

```python
trainx_rev.head()
```

```python
# model building
from sklearn.linear_model import LinearRegression
```

```python
lm=LinearRegression().fit(trainx_scaled,trainy)
```

```python
# prediction(predict)
testpre=lm.predict(testx_scaled)
```

```python
print('predicted value is: ',testpre[0:5])
print('actual value is: ',testy[0:5])
```

```python
# model evalutation
from sklearn.metrics import mean_squared_error,mean_absolute_percentage_error
```

```python
mean_squared_error(testpre,testy)
```

```python
mean_absolute_percentage_error(testpre,testy) # 14% error
```
